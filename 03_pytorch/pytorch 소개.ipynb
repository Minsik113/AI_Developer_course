## 2-1. Tensor
- Numpy: ndarray-list
- PyTorch: Tensor-list를 이용함.
- 이 둘이 거의 동일하다~

* 기본적으로 tensor가 가질수 있는 data 타입은 numpy와 동일


* 기본적으로 Numpy의 대부분의 사용법이 그대로 적용됨
    - x_data[1:]
    - x_data[:2, 1:]
    - x_data.flatten()
    - x_data.torch.ones_like(x_data) # 동일한 크기의 행렬만들어줌. 자주 쓰임
    - x_data.shape
    - x_data.dtype

- pytorch의 tensor는 GPU에 올려서 사용가능

* view, squeeze, unsqueeze 등으로 tensor 조정가능
    - view: reshape과 동일하게 tensor의 shape을 변환
        - view랑 reshape이 조금 다르다. 나중에 배움. 모르겠으면 view쓰면됨
    - squeeze: 차원의 개수가 1인 차원을 삭제 (압축)
    - unsqueeze: 차원의 개수가 1인 차원을 추가

## 2-3. Tensor operations

####  기본적인 tensor의 operations는 numpy와 동일

#### 행렬곱셈 연산은 함수는 dot이 아닌 mm 사용

    - pytorch에서 내적을 구하는 dot함수와 행렬곱을 구하는 mm을 구분해서 쓴다.


#### mm과 matmul은 broadcasting 지원 차리
- matmul에서 (5,2,3)에서 5는 batch이고 2 by 3 이다.
- 그아래 3 by 1과의 연산이다
- 그래서 5x2x1이됨
