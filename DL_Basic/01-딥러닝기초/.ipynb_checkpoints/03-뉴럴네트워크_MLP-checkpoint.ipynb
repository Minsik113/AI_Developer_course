## 1.Neural Networks
- Neural networks are function approximators that stack affine transformations followed by nonlinear transformations.
- 어떤함수를 모방하는 function approximators

## 2.Linear Neural Networks
- 선형회귀
- 입력이 1차원이고 출력이 1차원일때 이 두개를 연결하는 모델을 찾는 것
- 기울기와 y절편을 찾는 문제

- x에서 y햇으로 가는 모델. 
- loss는 우리가 이루고자 하는 현상이 이루어졌을때 줄어는 함수.
- 이 선형모델을 만드는 w와 b를 찾는 문제
- 내가 가진 N개의 데이터를 잘 설명하거나 표현할 수 있는 모델을 찾는거니까 뉴럴네트워크의 출력값과 나의 데이터의 차이를 줄이는. 이 차이를 정의할때 square loss(제곱)을 사용하고 이걸 MSE라함.
- 파라미터를 어디로 움직였을 때 loss function이 줄어드는지를 확인해야함
    - 미분해서 지금나온 방향의 음수방향으로 업데이트하게되면 loss 가 최소화되는 지점에 이르게 되겠지~
- loss에 대해 w와 b 각각의 편미분을 적용. w와 b를 계속 업데이트 해나가는게 gradient descent(경사하강법)라고 한다.
- 적잘한 stepsize를 잡는게 가장 중요한 문제중 하나이다. 
    - adaptive learning rate은 이 lr을 자동적으로 조절해줌

#### x데이터를 W와 b(bias)를 통해 출력으로 보내겠다.

1. 데이터를 W1와 곱해서 나온게 h(히든벡터)라고 한다. (바이어스는 빼고 이야기)
2. h에 W2를 곱해서 층 늘려도된다. 하지만 이렇게하면 단순히 행렬 2개를 곱한거랑 같다.
3. 그래서 우리는 nonlinear transform이 필요하다.
    - h에 nonlinear transform을 해줘야한다 Relu나 tanh와같은 activation function을 곱해줘야한다.
    
#### 정리하자면
1. x라는 입력에서 y라는 출력으로 가는 이 맵핑이 표현할 수 있는 표현력을 최대한 극대화하기 위해서는 단순히 선형결합을 N번 반복하는게 아니라, 한번 선형결합이 반복되게 되면 그 뒤에 activation function을 곱해줘여한다.(sigmoid, tanh등) 
2. nonlinear transform을 거치고, 이렇게 얻어진 feature vector를 다시 선형변환을한다. 
3. 이결과를 다시 nonlinear transform을 거친다. n번 반복하게 되면 더 많은 표현력을 갖게된다.
    

-> 즉, 어떤 activation function이 좋은지는 문제마다 다르다. 하지만 확실한건 이러한 nonlinear transform이 들어가야 깊게 쌓았을 때, 더 많은 표현력을 가질수 있으므로 의미가 있다.

- 히든레이어가 1개만있는 뉴럴네트워크의 표현력은 우리가 일반적으로 생각하는 continuous function들을 다 포함한다. 그렇기때문에 잘된다. 이런걸 만족시키는 뉴럴네트워크가 존재한다는 의미이지, 내가 만든 모델이 항상 좋다는건 아님.

-> 그만큼 뉴럴네트워크의 표현력이 크다!

# 3. Loss function
#### Regression Task
- MSE - 제곱을 이용
- 에러가 많이꼈음. 이 에러가 타겟데이터와 엄청나게 큰 차이가 있다고하자. 이런경우 이 데이터를 맞추려다가 뉴럴네트워크가 전반적으로 망가지게 된다. 그래서 이 L2 norm은 항상 도움이 되는것은 아니다. (L1 norm 은 절대값을 다 더하는 것이었음)
#### classification Task
- 분류문제는 보통 one hot vectior로 표현된다.
- d개의 라벨을 갖는 분류문제를 풀때, d개의 아웃풋중에서 제일 큰 출력을 가진 index만을 고려한다. 그 값이 100이든 10이든 중요하지않음. 어떤게 가장 유력한지가 중요하다. 이것을 수학적으로 표현하기 까다로우니까 분류문제를 풀때는 CE(Cross Entropy)를 사용하는거다.
#### Probabilistic Task
- 출력값이 숫자가 아니라 확률적인 문제를 하고 싶을 때.
- ex) 사람의 얼굴을 보고 나이를 맞추고 싶다. 
- 가우시안의 log likelihood를 maximize하겠다 해서 Maximum likelihood estimation이다
