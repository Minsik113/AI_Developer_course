# 4-2. ResNet 레즈넷
- 이 연구 덕분에 딥러닝이 딥러닝이 가능해졌다.
    - 딥러닝이 "뉴럴네트워크를 활용하는 어떤 기계학습 분야"라고 했지만, 왜 딥러닝이냐? 라는 질문에 설명을 하지 않았다. 설명할 수 없으니까.
    - 딥러닝은 뉴럴네트워크를 깊게 쌓는거. 네트워크를 깊게 쌓으면 학습이 잘 안됨. train이 잘 안된다는게 아니라 test할때 성능이 좋게 안나온다는 의미.
    - 층을 어느정도 깊게 쌓으면 성능이 더 좋아지지 않음
- 하지만 resnet이 나온 후로 변함. 20층까지 쌓으면 성능이 더이상 안좋아지던게 100층까지 쌓아도 test데이터의 성능이 좋게 만들어준 굉장히 좋은 논문

# 6. Bert 벌트
- Bidirectional Encoder Representations from Transformers
- 앞에서 본 transformer구조를 활용하는데 bidirectional encoder를 활용하는 기술
- fine-tuning모델이 중요하다
    - 자연어 모델. 즉, NLP 모델은 보통 language 모델을 학습한다.
    - language모델은 이전의 단어를 보고, 다음에 어떤 단어가 나올지 맞추는거. 이를 반복하다보면 그럴싸한 문장이 만들어지고 그럴싸한 글이 써짐
- '내일의 날씨를 예측하는' 혹은 '좋은 뉴스기사를 작성하는' 네트워크를 만들고싶다면, 뉴스데이터가 필요하다. 뉴스데이터는 생각보다 많지않음. 하지만 일반적인 문장들이 많다. fine-Tuning모델은 굉장히 큰 말뭉치들을 활용해서 pre-training을 하는거다. 그 후, 내가 정말 풀고자하는 소수의 데이터를 fine-tuning하는 것이다. 그래서 좋은 성능을 내겠다 하는게 bert이다.

# 8. Self-Supervised Learning
- 2020년도의 트렌드를 꼽으라면 self-supervised learning이다.
- simCLR [심clr] 논문
- 이미지 분류와 같은 분류문제를 풀고 싶은데, 학습데이터는 한정적이다. 한정된 학습데이터로 모델,loss function을 바꿔가면서 여러가지 다른값들을 줘서 좋은 결과를 내는게 일반적인 방법이었다.
- 하지만, Self-Supervised Learning에서는 주어진 학습데인터 외에 라벨을 모르는 unsupervised data를 활용하겠다는 것이다.
    - 강아지, 고양이라고 되어있는 이미지 10만장
    - 라벨은 모르지만 굉장히 많은 이미지들을 학습에 같이 활용하고 싶다.
    - 1. 어떻게하면 이 이미지들을 컴퓨터가 잘 이해할 수 있는 벡터로 변형시킬지
    - 2. 단순히 학습데이터만 가지고 하는게 아니라 라벨이 없는 데이터들로 비지도학습을 함께 활용하면서 문제를 더 잘 풀겠다.
- simCLR은 이미지가 주어졌을때, 이 이미지에 대한 좋은 representation을 학습데이터 외의 데이터를 통해 추가적으로 얻겠다.
- 학습데이터를 도메인지식이나 가지고있는 시뮬레이터를 통해서 추가로 만들어내서 학습데이터를 뻥튀기시켜서 더 좋은 모델을 만들겠다 < 이쪽도 연구가 활발
