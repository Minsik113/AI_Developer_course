# 딥러닝이란?
1. 인간의 지능을 모방한다
2. 강아지인지 고양이인지 구분하는게 머신러닝이라면
3. 그 모델안에 세부적으로 딥러닝을 이용하는거.

-> 뉴럴네트워크를 사용하여 데이터를 학습하는거

### 논문을 볼 때, 이 4가지를 주의깊게 보면 이해하기 좋을거다.
1. 데이터를 학습할 모델
2. 데이터를 변형시킬 모델 (이미지넣으면 강아지인지 고양이인지 라벨이 나와야함)
3. 문제에따라 모델을 학습시키기위한 loss function
4. 이러한 loss function을 최소화 시키기위한 알고리즘(최근에는 r adam)

# Loss
1. 회귀문제의 경우, 실제값과 예측값의 차이제곱인 Mean Square Error를 주로사용
2. 분류문제의 경우, 실제값과 라벨값의 사이의 crossentropy를 최소로하는거 주로사용
3. 값에대한 평균과 분산 가우시안으로 모델링 하려할때, maximum likelihood estimation관점으로 문제를 풀 수 있다.
