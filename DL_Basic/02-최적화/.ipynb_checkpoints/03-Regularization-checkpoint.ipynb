# 1. Early stopping
- 학습을 멈추는거. 
- k-fold validation을 활용하든 지금까지 학습된 모델의 성능을 평가해보고(loss를 보고) 커지기 시작하려는 시점에서 멈춘다. 
- 빨리멈추자!

# 2. Parameter norm penalty
- 네트워크를 학습할때, 네트워크 weight의 숫자들이 작으면 작을수록 좋다. 크기관점에서.
- NN이 만들어내는 함수의 공간속에서 부드러운 함수를 만드려고
- 깔려있는 가정, 부드러운 함수일수록 generalize performance가 높을 것이다.

# 3. Data Augmentation
- 데이터가 무한히 많으면 다 잘된다.
- 데이터 수가 적으면 잘안됨.
- data augmentaion은 데이터 수를 늘리고 싶다는 것이다.
- 이미지를 옆으로 돌리던 조금씩 돌렸을때도 강아지로 인식하게해서 데이터 늘리는거.
- 숫자의 경우는, 이러면 label이 달라질 수 있으니 하면 안되겠지.
- 레이블이 변경되지 않는 한에서 data augmentation은 좋다.

# 4. Noise Robustness
- 입력데이터에 노이즈데잍터를 막 집어넣음
- 노이즈를 이미지뿐만아니라 weight에도 넣는다.
- 노이즈를 중간중간 집어넣게되면 그 네트워크가 성능이 더 잘 나온다 라는 실험적인 결과가 나온다.

# 5. Label smoothing
- 학습데이터를 하나만 보는게 아니라, 학습데이터 2개를 막 섞어줌.
- 2개의 클래스를 잘 구분할 수 있는 decision bound를 찾아서 분류를 잘 되게하는게 목적이다.
- 이러한 decision bound를 부드럽게 만들어줌
    - Mixup - 2개의 이미지를 고른 후 라벨을 섞는다. 이미지도 섞는다. 이미지가 고양이반 강아지반임. 그러면 결과는 고양이나 강아지가아니라, 고양이0.5 강아지0.5가 나오게 하는거.
    - Cutout - 이미지 주어지면 일정영역을 빼버린다
    - CutMix - 섞어줄때, 특정부분에 다른 이미지를 넣어줌
- Mixup의 경우는 성능이 많이 많이 올라간다.
- 분류문제를 풀고, 데이터셋이 한정적이고 더이상 얻을 수 없다면, mixup이나 cutmix활용해보면 좋다.
- 코드도 간단. 노력대비 성능좋음

# 7. Batch  Normalization (BN)
- Batch normalization compute the empirical mean and variance independently for each dimension (layers) and normalize.
- 내가 BN을 적용하려는 레이어에 statistic를 정교화 시키는 것이다. 
- 각각의 Layer가 
- 즉, 원래의 값들이 다 100정도를 가지면 0정도로 줄여버린다.
    - 평균을 빼주고, 표준편차를 나눠주고.
    - 논문에서는 internal Covariate shift를 줄인다(?)고 한다. Covariate는 feature랑 같은말.
    - 이후 논문들은 이걸 옹호하는 편은아님
- batch normalization을 활용하게 되면 일반적으로, layer가 깊게 쌓아지게 되면 성능이 높아진다.

### There are different variances of normalizations.
- batch norm은 layer전체에 대한 것을 줄였다면, 
- layer norm은 각각의 layer정보를 줄임
- instance norm은 데이터 1장별로 statistics를 정규화, 
- group norm은 layer norm과 instance norm의 중간

> 간단한 분류문제를 풀 때, batch noramlization이 성능을 높인다고 알려져있음
