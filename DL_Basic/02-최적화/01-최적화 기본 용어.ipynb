# 1. Generalization

# 2. Under-fitting vs. over-fitting 
- 과대적합: 학습데이터는 잘 맞추는데 실제는 잘 못맞춤
- 과소적합: 네트워크가 너무 간단하거나 training을 너무 조금해서 

# 3. Cross validation(교차검증)
- 학습데이터가 적거나 모델자체가 별로일 경우 성능이 낮을 수 있다. 이런걸 검증하기 위한게 cross validation이다.
- 학습데이터를 k개로 나눠서, 학습과 검증을 번갈아한다.
    - 5개로 나눴다면 1차는 1~4번까지 학습하고 5번을 테스트
    - 2차는 1,2,3,5로 학습하고 4번을 테스트
    - ...
    - 하이퍼 파라미터에 대한 어떤게 좋고 안좋은지 정의가 없으니, Cross validatio을 해서 최적의 하이퍼파리미터셋을 찾고, 이걸 고정하고 학습할때는 모든 데이터를 다 사용한다. 그래야 많이 할 수 있음. 테스트데이터는 절대 학습에 들어가선 안된다. 
    - 학습에는 train과 validation 데이터만 사용되어야하고, test데이터는 들어가면 절대안된다.
    

# 4. Bias-variance tradeoff 
- variance는 비슷한 입력을 넣었을 떄 출력이 얼마나 일관적인지를 의미한다. 사격할 때, 영점조절 같은거로 생각하면 됨.
    - variance가 큰 모델은 비슷한 입력이 들어와도 출력이 많이 달라진다. 오버피팅될 가능성이 크다.
- bias는 평균적으로 봤을 때, 출력이 분산이 크더라도 평균적으로 봤을때 얘네들이 다 true target에 접근하게 되면 bias가 낮다고한다
    - bias가 높다면 원하는 target과 거리가 멀다

-> cost는 bias, variance, noise 3가지로 나뉜다는 말이다. 모델의 특성이 bias를 많이 줄이게되면 variance가 높아지게 될 가능이 크고, variance가 높이게되면 bias가 줄어들 가능성이 크다.

- 근본적으로 학습데이터에 noise가 껴있는경우 bias랑 variance를 둘 다 줄일 수 있는 경우는 드물다. ? 

# 5. Bootstrapping 
- 부츠의 끈.
- Bootstrapping is any test or metric that uses random sampling with replacement.
- 기본적으로 학습데이터가 고정되어 있을때, 학습데이터 여러개를 만들고 여러모델을 만든다.
    - 데이터 100개가 있다면 80개를 학습데이터로 한다는 말.

# 6. Bagging and boosting
- 배깅(bagging)
    - 학습데이터를 여러개를 만들어서 bootstrapping하겠다는 말
    - 여러 모델들을 만들고 여러 output이 나오면 이들을 평균내겠다
    - =>앙상블
    - 80개 데이터로 n개의 모델을 만들고 이들의 평균을 냄. 
    > 즉, 데이터들을 부트스트랩핑으로 여러개 만들어서 여러 모델을 만들고, 각각의 모델을 독립적으로 돌려서 무언가 결과를 뽑겠다.
    
- 부스팅(boosting)
    - 모델을 간단하게 만들고, 학습데이터를 돌려본다.
        - 80개중 60개 잘맞추고, 20개 학습을 잘 못했다. 
    - 이번엔 학습을 잘 못했던 20개를 잘 학습할 수 있는 모델을 돌려본다
    - 하나하나의 모델들(weak learner)을 합쳐서 Sequence한 learner를 만든다. 각각의 weight를 찾아 합치는 느낌.
    > 즉, 여러개의 모델이 독립적이게 돌아가는게 아니라, 여러 모델들이 하나의 strong한 모델을 만들어내겠다.
