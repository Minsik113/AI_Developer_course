# Gradient Descent Methods
1. Stochastic gradient descent(SGD)
    - Update with the gradient computed from a single sample.
    - 10만개의 데이터가 있다면 1개에대해 구하고 업데이트하고 이걸 계속 반복하는거
    
2. Mini-batch gradient descent(MGD)
    - Update with the gradient computed from a subset of data.
    - 10만개나 1개가 아니라 batchsize를 설정해서 하나의 gradient를 구하고, 또 다른batch에서 업데이트하는거
    
3. Batch gradient descent
    - Update with the gradient computed from the whole data
    - 한번에 데이터 다 사용하겠다. 10만개 모두의 평균을 사용해 적용하겟다
   
-> 대부분은 mini batch를 사용한다

4. Gradient Descent Methods
    - (Stochastic) gradient descent 
    - gradient descent의 단점을 보완하는게 쭉 나온다.
        - Momentum
        - Nesterov accelerated gradient 
        - Adagrad
        - Adadelta
        - RMSprop
        - Adam

## Batch-size
- 64, 128 ... 이렇게 적용. 이 숫자는 굉장히 중요하다. 
- batch size가 크게쓰면 sharp minimizers
- batch size를 작게쓰면 flat minimizer

-> sharp minimizer보다 flat minimizer이 낫다.

>batch size를 줄이게 되면 generaize performance가 좋아진다는 걸 실험적으로 보여줌. 그렇다면 large batch size를 사용하려면 어떤 테크닉을 사용해야하는지 나오는 논문

# 2. Momentum
- 이전에 사용한 batch size에 해당하는 결과를 다음 결과에도 영향이 있게해보자.
    - 예를들어) n차시행에서 오른쪽으로 영향을 끼침. n+1차에서는 오른쪽으로 영향이 안미칠 수도 있지만, n+1차는 n차의 영향이 조금 있게 해보자.

> 모멘텀과 g라는 현재gradient가 들어가있는 a라는 값을 계속 가져간다. 
모멘텀과 현재 gradient를 합친 accumulation. 모멘텀이 포함된 gradient로 업데이트를 계속 시킨다. 장점은 한번 흘러가기 시작한 방향을 어느정도 유지시켜준다. 어느정도 잘 학습이 된 효과른 낸다.

# 3. Nesterov accelerated gradient 
- gradient랑 비슷. 
- Lookahead gradient를 계산
- 모멘텀은 현재 주어진 파라미터를 이용해 gradient를 계산해서, 모멘텀을 accumulation함.
- NAG는 모멘텀과 비슷하지만 한번 이동한다. 
    - a라고 불리는 현재 이동정보가 있다면 한번 가보고, 간곳에서 gradient를 가지고 accumulaion을 계산한다.
- 장점은 봉우리에 모멘텀보다 빨리 도달


# 4. Adagrad
- Adagrad adapts the learning rate, performing larger updates for infrequent and smaller updates for frequent parameters
- 뉴럴네트워크가 많이 변한 파라미터에 대해서는 조금 적게 변화시켜보고, 조금변한 파라미터는 많이 변화시킨다. 그렇다면, 지금까지 지나간 gradient를 다 저장해야겠네? 이게 공식의 G이다.
- adaptive learning 
- 단점은 G가 계속 커지기때문에, 결국 무한대로 가므로 W가 업데이트가 안된다. 뒤로가면갈수록 학습이 덜된다는 의미이다. -> 이를 해결하기 위한게 여러개있다. ex) Adam

# 5. Adadelta
- Adadelta extends Adagrad to reduce its monotonically decreasing the learning rate by restricting the accumulation window.
- Adagrad에서 파라미터가 엄청 많으면 G가 너무 커질것이다. 이를 막는 방법이 Exponentially Weighted Averages 이다.
    - 오래된 데이터의 영향과 최신 데이터의 영향이 비슷해져버리는 것을 막기위해 Weight를 부여하는 방식이 도입하였는데, 시간이 흐름에 따라 지수적으로 감쇠하도록 성계한 것이 지수가중평균(Exponentially Weighted Averages)이다.
- 즉, Exponentially Weighted Averages를 이용해 업데이트 하는게 Adadelta. 많이사용하진않음. 아래에 RMSprop를 더 사용

# 6. RMSprop (RMS 프랍)
- RMSprop is an unpublished, adaptive learning rate method proposed by Geoff Hinton in his lecture.
- Adadelta와 마찬가지로 G(gradient square)를 그냥 더해주지않고, 에타라는 step size를 넣어준다.
- 잘됐다!

# 7. Adam
- Adaptive Moment Estimation (Adam) leverages both past gradients and squared gradients.
- 일반적으로 가장 무난하게 사용하는게 아담
- G와 모멘텀을 함께 잘 섞어 사용한다.
