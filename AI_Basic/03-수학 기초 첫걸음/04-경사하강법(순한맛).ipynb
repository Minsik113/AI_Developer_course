# 1. 미분
- 미분(differentiation)은 변수의 움직임에 따른 함수값의 변화를 측정하기 위한 도구로 최적화에 제일 많이 사용되는 기법이다.


### 1-1. sympy[심파이]

```python
import sympy as sym
from sympy.abc import x
sym.diff(sym.poly(x**2 + 2*x + 3), x)
```

### 1-2. 미분은 함수 f의 주어진 점 (x,f(x))에서의 접선의 기울기를 구한다.

### 1-3. 한 점에서 접선의 기울기를 알면 어느 방향으로 점을 움직여야 함수값이 <font color='red'>증가</font>하는지/ <font color="blue">감소</font>하는지를 알 수 있다.

- <font color="red">미분값을 더하면 경사상승법(gradient ascnet)</font>라 하며 함수의 <font color="red">극대값</font>의 위치를 구할 때 사용한다. 목적함수를 <font color="red">최대화</font> 할 떄 사용

- <font color="blue">미분값을 뺴면 경사하강법(gradient ascnet)</font>라 하며 함수의 <font color="blue">극소값</font>의 위치를 구할 때 사용한다. 목적함수를 <font color="blue">최소화</font> 할 떄 사용

- <font color="red">경사상승</font>/ <font color="blue">경사하강</font> 방법은 극값에 도달하면 움직임을 멈춘다.

- lr은 학습률로서 미분을 통해 업데이트하는 속도를 조절한다.
    - 아무값을 넣는게 아니라 조심히 선정해야함
- 미분값을 계속 업데이트한다.

### 1-4. 변수가 벡터라면?
- 벡터가 입력인 다변수 함수의 경우 편미분(partial differentitaion)을 사용한다.
- 각 변수 별로 편미분을 계산한 그레디언트(gradient) 벡터를 이용하여 경사하강/경사상승법에 사용할 수 있다.


```python
import sympy as sym
from sympy.abc import x, y
sym.diff(sym.poly(x**2 + 2*x*y +3) + sym.cos(x + 2*y), x)
```

- 어느점이든 마이너스 그레디언트f는 극소점으로 움직이는 벡터가 그려진다.
