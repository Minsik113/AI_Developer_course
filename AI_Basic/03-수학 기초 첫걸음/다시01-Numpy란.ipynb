## 1. Numpy 특징
- 일반 List에 비해 빠르고, 메모리 효율적
- 반복문 없이 데이터 배열에 대한 처리를 지원함
- 선형대수와 관련된 다양한 기능을 제공함
- C, C++, 포트란 등의 언어와 통합 가능

- numpy는 np.array 함수를 활용하여 배열을 생성함 è ndarray
- numpy는 하나의 데이터 type만 배열에 넣을 수 있음
- List와 가장 큰 차이점, Dynamic typing not supported
    - 파이썬의 특징인 dynamic typing인데 np.array는 아니다. 메모리의 공간을 선언할때 잡아두는게아니라 추가하면서 쓸 수 있다. 하지만 numpy는 그렇지 않다.
- C의 Array를 사용하여 배열을 생성함

```python
import numpy as np
test_array = np.array([1,4,5,8], float)
print(test_array)
type(test_array[3])
```

```python
import copy
test_array1 = copy.deepcopy(test_array)
test_array1 is test_array
```

파이썬은 주소를 저장하는데, numpy는 값을 저장해서 훨씬빠르다.
