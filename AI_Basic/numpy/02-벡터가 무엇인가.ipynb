### 1-10. 내적은 정사영된 벡터의 길이와 관련이 있다.


* x내적y = x스칼라 * y스칼라 * cos(세타)

* 즉, 내적은 x벡터를 y벡터위로 정사영한 후, y길이만큼 조정한 값이다
    * x내적y = (x스칼라 * cos세타) * y스칼라
