# 1.Handling Shape

# 1-1. reshape
- Array의 shape의 크기를 변경함.
- element의 개수는 동일(데이터 개수 동일)


```python
import numpy as np
test_matrix = [[1,2,3,4]
              ,[1,2,5,6]]
np.array(test_matrix).shape # 2x4
print(f'np.array(test_matrix).reshape(8,) : {np.array(test_matrix).reshape(8,)}')
print(f'np.array(test_matrix).reshape(8,).shape : {np.array(test_matrix).reshape(8,).shape}')
```

```python
np.array(test_matrix).reshape(2,4)
```

- -1: size를 기반으로 row개수를 선정한다.
    - 뒤를 써줌으로써 앞은 자동으로 맞춰줌

```python
np.array(test_matrix).reshape(1,-1)
```

```python
np.array(test_matrix).reshape(-1,1)
```

```python
np.array(test_matrix).reshape(-1,2)
```

```python
np.array(test_matrix).reshape(2,2,2)
```

```python
np.array(test_matrix).reshape(2,-1,2)
```

# 1-6. ones, zerops and empty
- zeros: 0으로 가득찬 ndarray생성
    - np.zeros(shape, dtype, oreder)
- ones: 1으로 가득찬 ndarray생성
    - np.ones(shape, dtype, oreder)
- empty: shape만 주어지고 비어있는 ndarray생성
    - memory initialization이 되지 않음. -> 빈공간만 잡아준다.
    - 다른프로그램이 썻던 공간을 잡힐 수 있음. 이전 값들이 남아있을 수 있다.

```python
np.zeros(shape=(10,), dtype=np.int8) # 10 - zero vector 생성
```

```python
np.zeros(shape=(2,5), dtype=np.int8) # 10 - zero vector 생성
```

```python
np.ones(shape=(2,5), dtype=np.int8) 
```

```python
np.empty(shape=(2,5), dtype=np.int8)  # 실행시킬때마다 값이 바뀔것이다
```

```python
np.empty(shape=(2,5))
```

# 2-2. concatenate
- numpy array를 합치는(붙이는 함수)
- np.vstack
    - 아래로
- np.hstack
    - 옆으로
- np.concatenate
    - axis에 따라다름

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.vstack((a,b))
```

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
np.hstack((a,b))
```

```python
a = np.array([[1],[2],[3]])
b = np.array([[4],[5],[6]])
np.hstack((a,b))
```

```python
a = np.array([[1,2,3]])
b = np.array([[4,5,6]])
np.concatenate( (a,b), axis=0)
```

```python
a = np.array([[1,2,3]])
b = np.array([[4,5,6]])
np.concatenate( (a,b), axis=1)
```

```python
a = np.array([[1,2], [3,4]])
b = np.array([4,5])
np.concatenate( (a,b), axis=1) # 2x2랑 1x2니까 붙이려면 2x2로 만들어줘야한다.
```

```python
b= b[np.newaxis,:] # 
np.concatenate((a, b.T), axis=1) # 
```
