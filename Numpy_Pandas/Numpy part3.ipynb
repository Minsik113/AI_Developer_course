# 1-1. All & Any
- array의 데이터 전부(and) 또는 일부(or)가 조건에 만족 여부 반환
- np.logical_and
- np.logical_not
- np.logical_or

```python
import numpy as np
a = np.arange(10)
a
```

```python
a<4
```

```python
np.all(a<4)
```

```python
np.any(a<4)
```

```python
a = np.array([1,3,5], float)
b = np.array([2,4,6], float)
a < b
```

```python
(a<b).any()
```

```python
np.logical_and(a>0, a<3) # and 조건의 condition
```

```python
np.logical_not(np.array([True,False]))
```

# 1-3. argmax, argmin
- array내 최대값 또는 최소값의 index를 반환함
- axis기반의 반환

```python
a = np.array([1,2,3,5,7,8,1,231,-123])
np.argmax(a)
```

```python
a = np.array([[1,2,3,4],[5,6,7,8],[6,34,1,41]])
print(a)
# axis=1 -> axb에서 b임. 각행에서 최대값의 인덱스를 출력
np.argmax(a, axis=1), np.argmin(a,axis=0)
```

```python
a = np.array([1,2,3,5,7,8,1,231,-123])
a.argsort() # index값 보여줌
```

```python
a.argsort()[::-1]
```

```python
a[a.argsort()[::-1]]
```

# 2. boolean & fancy index

# 2-1.boolean index
- 특정 조건에 따른 값을 배열 형태로 추출
- comparison operation 함수들도 모두 사용가능

```python
test_array = np.array([1,4,0,2,3,8,9,7],float)
test_array > 3
```

```python
condition = test_array < 3
test_array[condition] # 조건이 True인 index element만 출력
```

# 2-2. fancy index
- numpy는 array를 index value로 사용해서 값 추출
- a.take(b)

```python
a = np.array([2,4,6,8], float)
b = np.array([0,0,1,3,2,1], int) # 반드시 integer로 선언
a[b] # bracket index, b배열의 값을 index로 하여 a값들을 추출
```

```python
a.take(b)
```

### boolean index는 boolean 배열을 쓰고 fancy index는 integer배열을 씀.

```python
a = np.array([[1,4],[9,16]],float)
b = np.array([0,0,1,1,0],int)
c = np.array([0,1,1,1,1],int)
a[b, c] # 0,0  0,1  1,1  1,1  0,1
```

```python
a = np.array([[1,4],[9,16]],float)
b = np.array([0,0,1,1,0],int)
a[b] # row값만 나온다.
```
