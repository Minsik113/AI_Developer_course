# 2. ndarray란
- 사실상 리스트랑 같은 개념이다. 다른점도 몇가지 있음
    - List랑 다르게 numpy는 하나의 데이터 type만 배열에 넣을 수 있다.
        - dynamic typing not supported
- np.array함수를 활용해 생성된 것을 ndarray[엔디어레이] 객체 라고한다. 

```python
import numpy as np
a = [1,2,3,4,5]
b = [5,4,3,2,1]

test_array = np.array(["1","2",5.0,6], float)
print(test_array) 
print(type(test_array[0])) # float
```

- 리스트는 값이 들어간게 아니라, 주소가 들어가있다.
- numpy array는 차례대로 값이 들어간다.
    - 그래서 메모리 크기가 일정하고, 그냥 값을 더하면 되는거라 속도도빠르고 메모리도 절약

## 2-1. List vs ndArray큰차이
- List 안에 주소를 저장하니까 1이라는 값을 가리키는 주소가 들어가므로 a[0],  b[-1]이 동일하다

```python
a = [1,2,3,4,5]
b = [5,4,3,2,1]
a[0] is b[-1] 
```

- ndarray는 값 자체가 들어가니까, 메모리 비교를 하더라도 다르다.

```python
a = np.array(a)
b = np.array(b)
a[1] is b[-2]
```

# 2-3. array의 dtype
- 64bit는 어디까지 표현이 가능할까?
    - 2^64가 아니라 +- 2^63이다. 마이너스도 표현해야하니까

- numpy로 array를 선언할때, data type을 지정해 줄 수 있고, 이게 얼마만큼의 용량을 차지하는지 알아보자
    - float32니까 32bit -> 4bytes이다. 4bytes가 6개니까 6*4bytes=24bytes만큼의 용량차지

### ndim은 number of dimensions

```python
a = [[1,2,3]
    ,[4,5,6]
    ,[7,8,9]] 
a = np.array(a, float)
print(f'a.shape { a.shape}') # 3x3
print(f'a.ndim { a.ndim}') # 2차원
```
